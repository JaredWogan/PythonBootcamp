```python
# Write a program which draws 200 samples from a Gaussian distribution 
# (mean = 75, standard deviation = 5). Use the random.gauss function from the random library.

# Imagine these numbers represent final grades of students in a course.

# Tasks:
# a) Print how many students failed the course (grade < 50) and how many got an A (grade > 90). (1 point)
# b) Plot a histogram of the student grades. Label the x-axis "Grade" and the y-axis "Students". (2 point)

# As this is the final quiz in the course, 1 point will be assigned for code cleanliness and readibility.

from typing import Union
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-pastel')

def simulateClass(
    mu: Union[int, float] = 75,
    sigma: Union[int, float] = 5,
    n: int = 200,
    binwidth: int = 5
) -> None:
    """Simulates a class of students where grades are taken from a Guassian distrubution
    mu: mean of the guassian distribution
    sigma: standard deviation of the guassian distribution
    n: number of students in the class
    binwidth: width of the bins in the histogram
    
    Returns: None
    """
    
    # Generate the random grades data    
    grades = np.random.normal(loc=mu, scale=sigma, size=n)
    grades[grades < 0] = 0
    grades[grades > 100] = 100

    # Count the number of students who failed and got an A
    failed = grades[grades < 50].shape[0]
    grade_A = grades[grades > 90].shape[0]
    
    print("Number of students who failed: ", failed)
    print("Number of students who got an A (>90%): ", grade_A)
    
    # Generate a histogram of the grades
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Create bins that go from the min to the max of the data, with a binwidth set by function parameters
    bins = bins=np.arange(grades.min(), grades.max() + binwidth, binwidth)
    ax.hist(grades, bins=bins, edgecolor='black')
    
    ax.set_title("Student Grades", fontsize=20)
    ax.set_xlabel("Grade", fontsize=16)
    ax.set_ylabel("Students", fontsize=16)

simulateClass(75, 5, 200)
simulateClass(75, 5, 1_000)
simulateClass(75, 5, 10_000)
```

```python
# Write a function which takes the mass and velocity as inputs and returns the kinetic energy computed
# using a classical relation KE = 1/2*m*v^2. (1 point)

# Compute and print out the energy in Joules of a rifle bullet (50 g, 1.2 km/s) (1 point) and a 
# mm-sized meteoroid (10^-5 g, 60 km/s) (1 point).

# As this is the final quiz in the course, 1 point will be assigned for code cleanliness and readibility.

from typing import Union
def kineticEnergy(mass: Union[int, float], velocity: Union[int, float]) -> Union[int, float]:
    """Computes the kinetic energy given the mass and velocity of an object in SI units.
    mass: int | float (kg)
    velocity: int | float (m/s)
    Returns: int | float  -> Kinetic Energy (J)
    """
    return 1/2 * mass * velocity**2

mass_bullet = 50 / 1000 # grams / 1000 = kg
velocity_bullet = 1.2 * 1000 # km/s * 1000 = m/s
print(f"Kinetic Energy of a rifle bullet: {kineticEnergy(mass_bullet, velocity_bullet)} J")

mass_meteoroid = 10**-5 / 1000 # grams / 1000 = kg
velocity_meteroid = 60 * 1000 # km/s * 1000 = m/s
print(f"Kinetic Energy of a mm-sized meteoroid: {kineticEnergy(mass_meteoroid, velocity_meteroid)} J")
```

```python
# Write a function which converts geographical coordinates from the sexagesimal 
# format (degrees, minutes, seconds) to decimal degrees. Each degree has 60 minutes 
# and each minute has 60 seconds. The function should take one variable as input, the 
# coordinates in the DDdMMmSSs format. For example, the input "25d34m59s" should return 25.58305555. (1 point)

# As this is the final quiz in the course, 1 point will be assigned for code cleanliness and readibility.

def sexagesimalToDecimal(input_string: str) -> float:
    """Converts sexagesimal coordinates to decimal degrees.
    sexagesimal: str -> "DDdMMmSSs"
    Returns: float -> Decimal Degrees
    """
    degrees, *r = input_string.split("d")
    minutes, *r = r[0].split("m")
    seconds, *r = r[0].split("s")
    return int(degrees) + int(minutes)/60 + int(seconds)/3600

print(sexagesimalToDecimal("25d34m59s"))
```
